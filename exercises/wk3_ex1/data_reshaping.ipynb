```python
#start by importing the pandas library
import pandas as pd
```

```python
#crating my dataframe in  wide format
hospital = pd.DataFrame ({
    "patient":["p001","p002"],
    "jan_BP":[120,130],
    "feb_BP":[125,135], 
    "mar_BP":[122,132]
})
#Here i print the hospital dataframe to see how it looks like in wide format
print(hospital)
```

```python
#transforming the data frame from wide to melted format
hospital_melted = hospital.melt(
    id_vars=["patient"], 
    var_name="month", 
    value_name="Blood Pressure")

# i now print the hospital_melted dataframe to see how it looks like in long format
print(hospital_melted)
```

```python
hospital_melted = pd.melt(
    hospital,
    id_vars=['patient'],
    var_name='Month',
    value_name='Blood_Pressure'
)

print(hospital_melted)
```

```python
#converting the long format back to wide format
hospital_wide = hospital_melted.pivot(
    index='patient',
    columns='Month',
    values='Blood_Pressure'
)

#printthe widw datframe again to see how it looks like
print(hospital_wide)
```

```python
#convert long format back to wide format and reset the index

hospital_wide = hospital_melted.pivot(
    index="patient",
    columns="Month",
    values="Blood_Pressure"
)

#convert the index to a normal column
hospital_wide=hospital_wide.reset_index()

print(hospital_wide)
```
