```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# create a dataframe
df = pd.DataFrame({
    "crop": ["Maize", "Rice", "Cassava", "Yam", "Beans", "Tomato", "Onion", "Pepper", "Carrot", "Cabbage", "Pawpaw", "Apple"],
    "price_ghs": [15.5, 20.0, 10.0, 25.0, 18.5, 12.0, 14.0, 18.0, 12.5, 17.0 , 11.0, 19.5],
    "region": ["Ashanti", "Volta", "Eastern", "Northern", "Ashanti", "Greater Accra", "Upper East" , "Northern" , "Greater Accra", "Ashanti", "Western", "Eastern"],
    "quantity_kg": [100, 150, 200, 120, 130, 90, 110, 220, 320, 340, 430, 440],
    "month": ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October","November", "December"]
})
```

```python
# Create subplots
fig,(ax1, ax2, ax3) = plt.subplots(1, 3, figsize= (18,6))

# ax1: Bar chart - total revenue per crop
ax1 = df.groupby("crop")["price_ghs"].sum().plot(kind="bar", ax=ax1)
# Set Title
ax1.set_title("Total revenue per crop")
# Set x-label
ax1.set_xlabel("Crop")
# Set Y-label
ax1.set_ylabel("Price (GHC)")
ax1.tick_params(axis="x", rotation=45)

# ax2: Line chart -  monthly avg price per region
df.groupby(["month", "region"])["price_ghs"].mean().unstack().plot(kind="line", ax=ax2)
# Set Title
ax2.set_title("Monthly Avg Price per Region")
# Set x-label
ax2.set_xlabel("Month")
# Set y-label
ax2.set_ylabel("Price (GHC)")
ax2.legend()

# ax3: Boxplot - price distribution by region
sns.boxplot(x="region", y="price_ghs", data=df, ax=ax3)
ax3.set_title("Price Distribution by Region")
plt.tight_layout()
# Save image as png
plt.savefig("crop_dashboard.png", dpi=150)
```

```python
import matplotlib.pyplot as plt

# Create a figure with a specific size
plt.figure(figsize=(10, 5))

# Create a bar chart: crops on x-axis, price on y-axis
plt.bar(df["crop"], df["price_ghs"])

# Add a title and labels
plt.title("Price of Crops")
plt.xlabel("Crop")
plt.ylabel("Price (GHS)")

# Rotate x-axis labels so they don't get squished
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
