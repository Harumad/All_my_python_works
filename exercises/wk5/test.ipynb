Importing messy data to be cleaned


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student_data.csv")
```

```python
df
```

```python
print(df["name"].tolist())
df["name"] = df["name"].str.strip().str.replace(r"\s+", " ", regex=True).str.title()
print(df["name"].tolist())
```

```python
df["gender"] = df["gender"].str.title()
print(df["gender"].tolist())
```

```python
df["class"] = df["class"].str.title()
print(df["class"])
```

```python
df.dtypes
```

```python
median_math_score = df["math_score"].median()
df["math_score"] = df["math_score"].fillna(median_math_score)
```

```python
median_english_score = df["english_score"].median()
df["english_score"] = df["english_score"].fillna(median_english_score)
```

```python
median_science_score = df["science_score"].median()
df["science_score"] = df["science_score"].fillna(median_science_score)

print(df["science_score"])
```

```python
median_histry_score = df["history_score"].median()
df["history_score"] = df["history_score"].fillna(median_histry_score)

print(df["history_score"])
```

```python
df["dob"] = df["dob"].str.replace("-", "/")
```

```python
df["tuition_paid"] = df["tuition_paid"].str.title()
print(df["tuition_paid"].tolist())
```

```python
df["contact"] = df["contact"].str.replace("-", "").str.replace(" ", "")
df["contact"] = df["contact"].fillna("0024377522")
# df["contact"] = df["contact"].astype(int)
df["contact"]
```

```python
df.dtypes
```

```python
df["address"] = df["address"].str.title() 

print(df["address"].tolist())
```

```python
df["house"] = df["house"].str.title()
print(df["house"].tolist())
```

```python
print(df)
```

```python
print(df[df["name"].duplicated(keep=False)])
df
```

```python
df.info()
```

```python
df["tuition_amount"] = df["tuition_amount"].str.replace("GHS", "").str.replace(",", "")
df["tuition_amount"] = df["tuition_amount"].astype(int)
df["tuition_amount"].dtypes
```

```python
df["dob"] = pd.to_datetime(df["dob"], dayfirst=True, errors="coerce")

df["dob"] = df["dob"].dt.strftime("%Y/%-m/%-d")

df["dob"] = df["dob"].fillna(pd.Timestamp("2000/01/01"))

df["dob"] = pd.to_datetime(df["dob"])

df.dtypes
```

```python
df["gender"] = df["gender"].replace({
    "M": "Male",
    "F": "Female"
})

df["gender"]
```

```python
# Remove duplicates
df = df.drop_duplicates(subset=["name"])
# Fix scores > 100
df.loc[df["math_score"] > 100, "math_score"] = 100
# Handle fake DOB
# df.loc[df["dob"] == "2000/01/01", "dob"] = None
# Clean contact numbers
df["contact"] = df["contact"].str.replace("00243", "0", regex=False)
# 5. Find missing/short names
df[df["name"].str.split().str.len() < 2]
# Check inconsistencies
df.groupby("address")["house"].unique()
```

```python
df.to_csv("clean.csv", index=False)
```
