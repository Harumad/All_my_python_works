```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clean_school_system.csv')

df
```

```python
class_order=['Grade 9','Grade 10', 'Grade 11']
avg_math=df.groupby('class')['math_score'].mean().round(2)
avg_math=avg_math.reindex(class_order)
print(avg_math)
```

```python
plt.figure(figsize=(8,5))
plt.title('Average Math Scores by Class')
plt.bar(avg_math.index,avg_math.values,color=['blue','orange']) 
plt.xlabel('Class') 
plt.ylabel('Average Math Score')
plt.show()
```

```python
city_counts=df['address'].value_counts()
print(city_counts)
```

```python
plt.figure(figsize=(8,5))
plt.barh(city_counts.index,city_counts.values,color=['blue','orange'])
plt.title('Number of Students by City')
plt.xlabel('City')
plt.ylabel('Number of Students')
plt.show()
```

```python
age_counts=df['age'].value_counts()
print(age_counts)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'])
print(plt.show())
```

```python
avg_english=df.groupby('class')['english_score'].mean().round(2)
print(avg_english)  
```

```python
plt.figure(figsize=(8,5))
plt.title('Average English Scores by Class')
plt.bar(avg_english.index,avg_english.values,color=['blue','orange'])
plt.xlabel('Class')
plt.ylabel('Average English Score')
plt.show()
```

```python
tuition_counts=df['tuition_amount'].value_counts()
print(tuition_counts)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['tuition_amount'])
plt.xlabel('Tuition Amount')
plt.ylabel('frequency')
plt.title('Distribution of Tuition Amounts')
plt.show()
```

```python
tuition_counts=df['tuition_paid'].value_counts()
print(tuition_counts)
```

```python
df.drop_duplicates(subset=['name'], keep='first')
```

```python
df['tuition_paid'] = df['tuition_paid'].str.strip().str.lower()
tuition_counts=df['tuition_paid'].value_counts()
print(tuition_counts)


```

```python
plt.figure(figsize=(8,5))
plt.pie(tuition_counts.values, labels=tuition_counts.index, autopct='%1.1f%%', colors=['blue', 'orange'])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(
    tuition_counts.index,      # x-axis
    tuition_counts.values,     # y-axis
    marker='o',                # dots
    linestyle='--'             # dotted/dashed line
)

plt.xlabel("Tuition Paid")
plt.ylabel("Count")
plt.title("Tuition Payment Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(tuition_counts.index, tuition_counts.values, color='green')
plt.xlabel('Tuition Paid')
plt.ylabel('Count') 
plt.title('Tuition Payment Distribution')
plt.show()
```
