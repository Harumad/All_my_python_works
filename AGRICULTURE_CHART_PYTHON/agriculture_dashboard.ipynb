```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "crop": [
        "Maize","Rice","Cassava","Yam",
        "Maize","Rice","Cassava","Yam",
        "Maize","Rice","Cassava","Yam"
    ],
    "price_ghs": [10,15,8,12, 11,14,9,13, 12,16,10,14],
    "region": [
        "Accra","Accra","Accra","Accra",
        "Kumasi","Kumasi","Kumasi","Kumasi",
        "Tamale","Tamale","Tamale","Tamale"
    ],
    "quantity_kg": [100,80,120,90, 110,85,130,95, 105,78,125,100],
    "month": [
        "Jan","Feb","Mar","Apr",
        "Jan","Feb","Mar","Apr",
        "Jan","Feb","Mar","Apr"
    ]
}

df = pd.DataFrame(data)
print(df)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,6))
```

```python
df["revenue"] = df["price_ghs"] * df["quantity_kg"]
revenue_per_crop = df.groupby("crop")["revenue"].sum()
revenue_per_crop = df.groupby("crop")["revenue"].sum()
```

```python
revenue_per_crop.plot(kind="bar", ax=ax1)

ax1.set_title("Total Revenue per Crop")
ax1.set_xlabel("Crop")
ax1.set_ylabel("Revenue (GHS)")
ax1.tick_params(axis='x', rotation=45)
```

```python
avg_price = df.groupby(["month","region"])["price_ghs"].mean().unstack()
avg_price.plot(ax=ax2)

ax2.set_title("Monthly Avg Price Trend")
ax2.set_xlabel("Month")
ax2.set_ylabel("Price (GHS)")
ax2.legend(title="Region")


sns.boxplot(data=df, x="region", y="price_ghs", ax=ax3)

ax3.set_title("Price Distribution by Region")
ax3.set_xlabel("Region")
ax3.set_ylabel("Price (GHS)")
```

```python
plt.tight_layout()
plt.savefig("crop_dashboard.png", dpi=150)
plt.show()
```
