```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("employees_clean.csv")

plt.figure(figsize=(10,5))
plt.scatter(df["age"], df["salary"], color="blue", s=20)

# Add labels (names)
for _, row in df.iterrows():
    plt.text(row["age"], row["salary"] + 300, row["name"].split()[0])

plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
